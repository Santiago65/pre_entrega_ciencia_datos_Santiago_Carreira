## üîπ 4- Introducci√≥n a Pandas: realizar un an√°lisis exploratorio inicial de los DataFrames.

```python
# ============================================
#  AN√ÅLISIS EXPLORATORIO INICIAL
# ============================================

# Analizamos estructura y contenido de cada DataFrame

def exploracion(df, nombre):
    print(f"\n### An√°lisis del DataFrame: {nombre}")
    display(df.head(3))
    print("\nInformaci√≥n general:")
    print(df.info())
    print("\nEstad√≠sticas descriptivas:")
    display(df.describe(include="all"))
    print("\nCantidad de filas y columnas:", df.shape)
    print("\nTipos de datos:")
    print(df.dtypes)
    print("-" * 60)

exploracion(ventas, "Ventas")
exploracion(clientes, "Clientes")
exploracion(marketing, "Marketing")



```

## üîπ 5- Calidad de Datos: Identificar valores nulos y duplicados en los conjuntos de datos.

```python
# ============================================
#  CALIDAD DE LOS DATOS
# ============================================

# Identificar valores nulos y duplicados

def calidad(df, nombre):
    print(f"\n### An√°lisis de calidad: {nombre}")
    print("Valores nulos por columna:")
    display(df.isna().sum().to_frame("Nulos"))
    print("Filas duplicadas:", df.duplicated().sum())

calidad(ventas, "Ventas")
calidad(clientes, "Clientes")
calidad(marketing, "Marketing")

```


## üîπ 6- **Limpieza del dataset**



```python

# ============================================
#  LIMPIEZA DE DATOS
# ============================================

# Crear copias para limpieza
ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

# Eliminar duplicados
ventas_clean.drop_duplicates(inplace=True)
clientes_clean.drop_duplicates(inplace=True)
marketing_clean.drop_duplicates(inplace=True)

# Limpieza de texto en columnas tipo string
for df in [ventas_clean, clientes_clean, marketing_clean]:
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].str.strip()

# Mostrar cambios realizados
print("Duplicados eliminados y textos normalizados.")
```

##üîπ 7- **Transformaci√≥n de datos**

```python
# ============================================
# TRANSFORMACI√ìN DE DATOS
# ============================================

# Clean and convert 'precio' to numeric
ventas_clean['precio'] = ventas_clean['precio'].astype(str).str.replace('$', '', regex=False)
ventas_clean['precio'] = pd.to_numeric(ventas_clean['precio'], errors='coerce')

# Create a column with the total of venta
ventas_clean["total_venta"] = ventas_clean["precio"] * ventas_clean["cantidad"]

# Calcular ventas totales por producto
ventas_por_producto = ventas_clean.groupby("producto")["total_venta"].sum().reset_index()

# Calcular el percentil 80
umbral_80 = ventas_por_producto["total_venta"].quantile(0.8)
print(f"Percentil 80 del total de ventas: {umbral_80:.2f}")

# Filtrar productos de alto rendimiento
productos_alto_rendimiento = ventas_por_producto.query("total_venta > @umbral_80")
print("\nProductos con alto rendimiento (superan el percentil 80):")
display(productos_alto_rendimiento)
```

##üîπ 8- **Agregaci√≥n**

```python
# ============================================
#  ETAPA 2.2: AGREGACI√ìN DE DATOS
# ============================================

# Agrupar ventas por categor√≠a y obtener m√©tricas
resumen_categorias = (
    ventas_clean.groupby("categoria")
    .agg(
        total_ingresos=("total_venta", "sum"),
        promedio_precio=("precio", "mean"),
        cantidad_vendida=("cantidad", "sum")
    )
    .reset_index()
)

print("Resumen de ventas por categor√≠a:")
display(resumen_categorias)
```

```python
ventas_clean.dropna(inplace=True)
print("Rows with null values dropped from ventas_clean.")
print("\nInformaci√≥n general despu√©s de eliminar nulos:")
display(ventas_clean.info())
print("\nValores nulos despu√©s de eliminar nulos:")
display(ventas_clean.isna().sum().to_frame("Nulos"))
```

##üîπ 9- **Integraci√≥n de datos**

```python
# ============================================
#  INTEGRACI√ìN DE DATOS
# ============================================

# Verificar columnas comunes entre ventas y marketing
columnas_comunes = set(ventas_clean.columns).intersection(set(marketing_clean.columns))
print("Columnas comunes entre ventas y marketing:", columnas_comunes)

# Si existe 'producto', realizar integraci√≥n
if "producto" in columnas_comunes:
    ventas_marketing = pd.merge(
        ventas_clean,
        marketing_clean,
        on="producto",
        how="left"
    )
    print("Integraci√≥n completada. Vista previa:")
    display(ventas_marketing.head())
else:
    print("No se encontr√≥ una columna com√∫n para integrar ambos datasets.")
```
